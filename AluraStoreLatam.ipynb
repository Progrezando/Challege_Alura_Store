#1. Análisis de facturación



```python
# Calculando los ingresos totales de cada tienda
ingreso_tienda1 = tienda['Precio'].sum()
ingreso_tienda2 = tienda2['Precio'].sum()
ingreso_tienda3 = tienda3['Precio'].sum()
ingreso_tienda4 = tienda4['Precio'].sum()
```

```python
# Mostrando los resultados
print(f"Ingresos Tienda 1: ${ingreso_tienda1:,.2f}")
print(f"Ingresos Tienda 2: ${ingreso_tienda2:,.2f}")
print(f"Ingresos Tienda 3: ${ingreso_tienda3:,.2f}")
print(f"Ingresos Tienda 4: ${ingreso_tienda4:,.2f}")
```

```python
import matplotlib.pyplot as plt

# 1. Organización de los datos
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
ingresos = [ingreso_tienda1, ingreso_tienda2, ingreso_tienda3, ingreso_tienda4]

# 2. Creación del gráfico de barras
plt.figure(figsize=(10, 6)) # Define el tamaño de la ventana
plt.bar(tiendas, ingresos, color=['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728'])

# 3. Títulos y etiquetas
plt.title('Ingresos Totales por Tienda - Alura Store', fontsize=15)
plt.xlabel('Tiendas', fontsize=12)
plt.ylabel('Ingresos (en miles de millones)', fontsize=12)

# 4. Mostramos el gráfico
plt.show()
```

# 2. Ventas por categoría

# 3. Calificación promedio de la tienda


```python
# Calculando el promedio de la valoración por tienda
promedio_t1 = tienda['Calificación'].mean()
promedio_t2 = tienda2['Calificación'].mean()
promedio_t3 = tienda3['Calificación'].mean()
promedio_t4 = tienda4['Calificación'].mean()
```

# 4. Productos más y menos vendidos

```python
# Función rápida para mostrar lo mejor y lo peor de cada tienda
def analizar_productos(df, nombre_tienda):
    top_vendas = df['Producto'].value_counts().head(3)
    peores_vendas = df['Producto'].value_counts().tail(3)

    print(f"--- {nombre_tienda} ---")
    print(f"MÁS VENDIDOS:\n{top_vendas}")
    print(f"MENOS VENDIDOS:\n{peores_vendas}\n")
```

```python
# Ejecutamos para las 4 tiendas
analizar_productos(tienda, "Tienda 1")
analizar_productos(tienda2, "Tienda 2")
analizar_productos(tienda3, "Tienda 3")
analizar_productos(tienda4, "Tienda 4")
```

```python
import matplotlib.pyplot as plt

# Configuramos la figura con 4 espacios (uno para cada tienda)
fig, axes = plt.subplots(2, 2, figsize=(15, 12))
fig.suptitle('Top 5 Productos Más Vendidos por Tienda', fontsize=20)

# Lista de nuestras tiendas y sus nombres para el bucle
mis_tiendas = [tienda, tienda2, tienda3, tienda4]
nombres = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
colores = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728']

# Bucle para crear los 4 gráficos de una vez
for i, ax in enumerate(axes.flat):
    datos_top = mis_tiendas[i]['Producto'].value_counts().head(5)
    datos_top.plot(kind='barh', ax=ax, color=colores[i])
    ax.set_title(nombres[i])
    ax.set_xlabel('Cantidad Vendida')
    ax.invert_yaxis() # Para que el #1 quede arriba

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

```python
# Resultados formateados
print(f"Costo de Envío Promedio Tienda 1: ${envio_t1:,.2f}")
print(f"Costo de Envío Promedio Tienda 2: ${envio_t2:,.2f}")
print(f"Costo de Envío Promedio Tienda 3: ${envio_t3:,.2f}")
print(f"Costo de Envío Promedio Tienda 4: ${envio_t4:,.2f}")
```

### 📄 Informe Final: Recomendación para Alura Store (Sr. Juan)

#### 1. Introducción

El presente análisis tiene como objetivo evaluar el rendimiento operativo y financiero de las cuatro sucursales de Alura Store. Se han cruzado variables de ingresos, costos de envío y calificación del cliente para identificar cuál de las tiendas presenta el modelo de negocio menos sostenible a largo plazo.

#### 2. Desarrollo del Análisis

Tras procesar los datos, se identificaron los siguientes patrones clave:

**Ingresos y Volumen**: La Tienda 1 lidera la facturación bruta ($1,150,880,400.00). Sin embargo, su volumen de ventas está concentrado en categorías pesadas (Muebles), lo que impacta directamente en los costos de envío.

**Eficiencia Logística**: Existe una correlación negativa entre el costo de envío y la calificación. La Tienda 1 presenta el costo de envío más alto ($26,018.61), lo que castiga el precio final pagado por el cliente.

Satisfacción del Cliente: A pesar de ser la que más factura, la Tienda 1 tiene la calificación más baja (3.97), indicando que el volumen de ventas no garantiza la fidelidad ni la calidad del servicio.

Fortalezas Comparativas: La Tienda 3 destaca por tener la mejor calificación (4.04), mientras que la Tienda 4 muestra la mayor eficiencia en costos de envío.

#### 3. Conclusión y Recomendación

Se recomienda al Sr. Juan proceder con la venta de la Tienda 1.

**Justificación**

**_Ineficiencia Logística_**: Es la tienda con los costos de envío más elevados, lo que sugiere una gestión de transporte deficiente o contratos de carga desfavorables.

**_Riesgo de Reputación_**: Posee la calificación más baja del grupo. En un mercado competitivo, la insatisfacción del cliente es un indicador temprano de una futura caída en las ventas.

**_Bajo Lucro Real Potencial_**: Aunque su facturación bruta es alta, el costo operativo derivado de los envíos y la gestión de categorías pesadas (muebles) reduce el margen neto en comparación con tiendas más eficientes como la 3 o la 4.
